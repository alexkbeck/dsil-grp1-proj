```python
# Get the highest adjusted close for each ticker
%%bigquery
SELECT
    MAX(Adj_Close) AS max_adj_close,
    Ticker
FROM
    stock_data.sp500
GROUP BY
    Ticker -- When you perform an aggregation like SUM() or MAX(), you must GROUP BY any other columns in your SELECT statement
ORDER BY
    MAX(Adj_Close) DESC -- This orders the results by the Adjusted Close in descending order
```

```python
# Get the adjusted close for each ticker on the first day of 2019
%%bigquery
SELECT
    Ticker,
    Adj_Close,
    Date
FROM
    stock_data.sp500
WHERE
    Date = (
            SELECT
                MIN(Date)
            FROM
                stock_data.sp500
    )
```

```python
# Join the two views above together to see the price on 2019-01-01 vs. the most recent price
%%bigquery
WITH            -- Use WITH temp_table_name AS (query) to set up a temporary table
    min_date    -- We do this here to save the first adj close for each ticker to a temporary table
    AS (
        SELECT
            Ticker,
            Adj_Close,
            Date
        FROM
            stock_data.sp500
        WHERE
            Date = (
                    SELECT
                        MIN(Date)
                    FROM
                        stock_data.sp500
            )
    )
SELECT  -- Now we select the data we want from the temporary table
    min_date.Ticker,
    min_date.Adj_Close AS first_adj_close,
    min_date.Date AS first_date,
    max_date.Adj_Close AS last_adj_close, -- Select data from the join that is used below
    max_date.Date AS last_date
FROM
    min_date
INNER JOIN (    -- INNER JOIN will combine data from different tables based on a linked column
        SELECT
            Ticker,
            Adj_Close,
            Date
        FROM
            stock_data.sp500
        WHERE
            Date = (
                    SELECT
                        MAX(Date)
                    FROM
                        stock_data.sp500
            )
) AS max_date   -- Use AS to create an alias for any table
    ON max_date.Ticker = min_date.Ticker    -- In this join we link on the Ticker symbol for both the min_date and max_date views
```

```python
# Same query as the cell above, but now we will calculate the % change in price for each ticker in a new column (new lines have comments)
%%bigquery
WITH
    min_date
    AS (
        SELECT
            Ticker,
            Adj_Close,
            Date
        FROM
            stock_data.sp500
        WHERE
            Date = (
                    SELECT
                        MIN(Date)
                    FROM
                        stock_data.sp500
            )
    )
SELECT
    min_date.Ticker,
    min_date.Adj_Close AS first_adj_close,
    min_date.Date AS first_date,
    max_date.Adj_Close AS last_adj_close,
    max_date.Date AS last_date,
   ( (max_date.Adj_Close - min_date.Adj_Close) / min_date.Adj_Close ) * 100 AS pct_change_in_price -- Here we calculate the percent change using the formula (x2 - x1) / x1
FROM
    min_date
INNER JOIN (
        SELECT
            Ticker,
            Adj_Close,
            Date
        FROM
            stock_data.sp500
        WHERE
            Date = (
                    SELECT
                        MAX(Date)
                    FROM
                        stock_data.sp500
            )
) AS max_date
    ON max_date.Ticker = min_date.Ticker
ORDER BY pct_change_in_price DESC   -- This will order the results by the best performing stock
LIMIT 10    -- Just see the best 10 stocks. Nvidia increased 436%!!!
```

```python
# Same query again, but this time filtering on Thermo Fisher Scientific
%%bigquery
WITH
    min_date
    AS (
        SELECT
            Ticker,
            Adj_Close,
            Date
        FROM
            stock_data.sp500
        WHERE
            Date = (
                    SELECT
                        MIN(Date)
                    FROM
                        stock_data.sp500
            )
    )
SELECT
    min_date.Ticker,
    min_date.Adj_Close AS first_adj_close,
    min_date.Date AS first_date,
    max_date.Adj_Close AS last_adj_close,
    max_date.Date AS last_date,
   ( (max_date.Adj_Close - min_date.Adj_Close) / min_date.Adj_Close ) * 100 AS pct_change_in_price
FROM
    min_date
INNER JOIN (
        SELECT
            Ticker,
            Adj_Close,
            Date
        FROM
            stock_data.sp500
        WHERE
            Date = (
                    SELECT
                        MAX(Date)
                    FROM
                        stock_data.sp500
            )
) AS max_date
    ON max_date.Ticker = min_date.Ticker
WHERE min_date.Ticker = 'TMO'   -- 138% increase, not bad!
ORDER BY pct_change_in_price DESC
```

```python
# Get the daily toal value for the S&P500
%%bigquery
SELECT
    Date,
    SUM(Adj_Close) AS total_value
FROM
    stock_data.sp500
GROUP BY
    Date
```

```python
# Get % change in daily value from the beginning of the data set until the end
%%bigquery
WITH
    first_total_value
    AS (
        SELECT
            Date,
            SUM(Adj_Close) AS total_value
        FROM
            stock_data.sp500
        WHERE
            Date = (
                    SELECT
                        MIN(Date)
                    FROM
                        stock_data.sp500
            )
        GROUP BY
            Date
    ),
    last_total_value
    AS (
        SELECT
            Date,
            SUM(Adj_Close) AS total_value
        FROM
            stock_data.sp500
        WHERE
            Date = (
                    SELECT
                        MAX(Date)
                    FROM
                        stock_data.sp500
            )
        GROUP BY
            Date
    )
SELECT
    first_total_value.total_value AS first_total_value,
    first_total_value.Date AS first_date,
    last_total_value.total_value AS last_total_value,
    last_total_value.Date AS last_date,
   ( (last_total_value.total_value - first_total_value.total_value) / first_total_value.total_value ) * 100 AS pct_change_in_price
FROM
    first_total_value, last_total_value
```

```python
# Total price per sector per day
%%bigquery
SELECT
    sp500.Date,
    fin.Sector,
    SUM(Adj_Close) AS sector_total_value
FROM
    stock_data.sp500 AS sp500
INNER JOIN 
        stock_data.fin AS fin 
            ON sp500.Ticker = fin.Symbol
GROUP BY
    sp500.Date,
    fin.Sector
```

```python
# Change in price per sector
%%bigquery
WITH
    sector_first_day_total_value
    AS (
        SELECT
            sp500.Date,
            fin.Sector,
            SUM(Adj_Close) AS sector_total_value
        FROM
            stock_data.sp500 AS sp500
        INNER JOIN 
                stock_data.fin AS fin 
                    ON sp500.Ticker = fin.Symbol
        WHERE
            Date = (
                    SELECT
                        MIN(Date)
                    FROM
                        stock_data.sp500
            )
        GROUP BY
            sp500.Date,
            fin.Sector
    ),
    sector_last_day_total_value
    AS (
        SELECT
            sp500.Date,
            fin.Sector,
            SUM(Adj_Close) AS sector_total_value
        FROM
            stock_data.sp500 AS sp500
        INNER JOIN 
                stock_data.fin AS fin 
                    ON sp500.Ticker = fin.Symbol
        WHERE
            Date = (
                    SELECT
                        MAX(Date)
                    FROM
                        stock_data.sp500
            )
        GROUP BY
            sp500.Date,
            fin.Sector
    )
SELECT
    sector_first_day_total_value.Sector,
    sector_first_day_total_value.Date AS first_date,
    sector_first_day_total_value.sector_total_value AS sector_first_day_value,
    sector_last_day_total_value.Date AS last_date,
    sector_last_day_total_value.sector_total_value AS sector_last_day_value,
    sector_last_day_total_value.sector_total_value - sector_first_day_total_value.sector_total_value AS sector_price_change,
    ((sector_last_day_total_value.sector_total_value - sector_first_day_total_value.sector_total_value) / sector_first_day_total_value.sector_total_value) * 100 AS sector_percent_change
FROM
    sector_first_day_total_value,
    sector_last_day_total_value
WHERE
    sector_first_day_total_value.Sector = sector_last_day_total_value.Sector
ORDER BY sector_percent_change DESC
```

```python
# Calculate "alpha" by sector
%%bigquery
DECLARE sp500_return FLOAT64 DEFAULT 84.17045537818937;
WITH
    sector_first_day_total_value
    AS (
        SELECT
            sp500.Date,
            fin.Sector,
            SUM(Adj_Close) AS sector_total_value
        FROM
            stock_data.sp500 AS sp500
        INNER JOIN 
                stock_data.fin AS fin 
                    ON sp500.Ticker = fin.Symbol
        WHERE
            Date = (
                    SELECT
                        MIN(Date)
                    FROM
                        stock_data.sp500
            )
        GROUP BY
            sp500.Date,
            fin.Sector
    ),
    sector_last_day_total_value
    AS (
        SELECT
            sp500.Date,
            fin.Sector,
            SUM(Adj_Close) AS sector_total_value
        FROM
            stock_data.sp500 AS sp500
        INNER JOIN 
                stock_data.fin AS fin 
                    ON sp500.Ticker = fin.Symbol
        WHERE
            Date = (
                    SELECT
                        MAX(Date)
                    FROM
                        stock_data.sp500
            )
        GROUP BY
            sp500.Date,
            fin.Sector
    ),
    return_by_sector
    AS (
    SELECT
        sector_first_day_total_value.Sector,
        sector_first_day_total_value.Date AS first_date,
        sector_first_day_total_value.sector_total_value AS sector_first_day_value,
        sector_last_day_total_value.Date AS last_date,
        sector_last_day_total_value.sector_total_value AS sector_last_day_value,
        sector_last_day_total_value.sector_total_value - sector_first_day_total_value.sector_total_value AS sector_price_change,
        ((sector_last_day_total_value.sector_total_value - sector_first_day_total_value.sector_total_value) / sector_first_day_total_value.sector_total_value) * 100 AS sector_percent_change

    FROM
        sector_first_day_total_value,
        sector_last_day_total_value
    WHERE
        sector_first_day_total_value.Sector = sector_last_day_total_value.Sector
    )
SELECT
    *,
    sp500_return,
    sector_percent_change - sp500_return AS sector_alpha
FROM
    return_by_sector
ORDER BY sector_percent_change DESC
```
